```python
from utils.evaluation import search_files
search_dir = "data/indian_villages/adj/"
pattern = ".allVillageRelationships_vil."
file_list = search_files(directory=search_dir, pattern=pattern)
```

```python
import numpy as np
import networkx as nx
from collections import defaultdict
from diffusion.jackson_metrics import diffusion_centrality
from diffusion.complex_diffusion import paralell_complex_path
graph_dict = defaultdict(list)

for c, file in enumerate(file_list):
    adj = np.genfromtxt(file, delimiter=",")
    G = nx.from_numpy_array(adj)
    graph_dict["size"].append(len(G))
    graph_dict["density"].append(nx.density(G))
    deg = nx.degree_centrality(G)
    deg_vals = np.fromiter(deg.values(), dtype=float)
    #deg_vals = deg_vals / deg_vals.max()
    graph_dict["degree_centrality"].append(deg_vals)
    #graph_dict["eigen_vector_centrality"].append(nx.eigenvector_centrality_numpy(G))
    dif = diffusion_centrality(G, T=5)
    vals = np.fromiter(dif.values(), dtype=float)
    #vals = vals / vals.max()
    graph_dict["diffusion_centrality"].append(vals)
    c = paralell_complex_path(G, T=4)
    c_vals = np.fromiter(c.values(), dtype=float)
    c_vals = c_vals / c_vals.max()
    graph_dict["complex_centrality"].append(c_vals)
    graph_dict["transitivity"].append(nx.transitivity(G))
    graph_dict["clustering"].append(nx.average_clustering(G))
    bc = nx.betweenness_centrality(G)
    graph_dict["betweenness_centrality"].append(np.fromiter(bc.values(), dtype=float))
```

```python
from tqdm import tqdm
from diffusion.jackson_metrics import Godfather
for c, file in enumerate(file_list):
    adj = np.genfromtxt(file, delimiter=",")
    G = nx.from_numpy_array(adj)
    god_dict = Godfather(G)
    god = np.fromiter(god_dict.values(), dtype=float)
    graph_dict["godfather_centrality"].append(god)
    graph_dict["god_gini"].append(calculate_gini(god))
    graph_dict["mean_god"].append(god.mean())
```

```python
import matplotlib.pyplot as plt
import itertools
from utils.evaluation import calculate_gini
for counter, deg in enumerate(graph_dict["degree_centrality"]):
    graph_dict["deg_gini"].append(calculate_gini(np.array(deg)))
    graph_dict["dif_gini"].append(calculate_gini(np.array(graph_dict["diffusion_centrality"][counter])))
    graph_dict["eigen_gini"].append(calculate_gini(np.array(graph_dict["eigenvector_centrality"][counter])))
    graph_dict["c_gini"].append(calculate_gini(np.array(graph_dict["complex_centrality"][counter])))
    graph_dict["b_gini"].append(calculate_gini(np.array(graph_dict["betweenness_centrality"][counter])))
    graph_dict["mean_degree"].append(np.array(deg).mean())
    graph_dict["mean_dif"].append(np.array(graph_dict["diffusion_centrality"][counter]).mean())
    graph_dict["mean_eigen"].append(np.array(graph_dict["eigenvector_centrality"][counter]).mean()) # probably hard to compare across different graphs
    graph_dict["mean_c"].append(np.array(graph_dict["complex_centrality"][counter]).mean())
    graph_dict["mean_b"].append(np.array(graph_dict["betweenness_centrality"][counter]).mean())
```
